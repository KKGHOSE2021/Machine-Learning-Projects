```python
# importing all required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv("Height-Weight Data.csv") # Load csv file
X = df[['Height']] # Take feature values
y = df['Weight'] # Take target values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # Split feature and target value for training and testing
```

```python
reg = linear_model.LinearRegression() # create object of linear regression
reg.fit(X_train, y_train) # Fit the linear regression by training values
y_pred = reg.predict(X_test) # Predict target values for test dataset using the trainned model 
```

```python
print("m = ", reg.coef_) # slope
print("c = ", reg.intercept_) # y-intercept
print("MSE = ", mean_squared_error(y_test, y_pred)) # The mean squared error
print("MAE =", mean_absolute_error(y_test, y_pred)) # The mean absolute error
print ("Traing Accuracy: ", reg.score(X_train, y_train)) # Check training accuracy
print ("Testing Accuracy: ", reg.score(X_test, y_test)) # Check testing accuracy
```

```python
# Comparing the generated linear model with the given dataset
plt.plot(df.Height, reg.predict(df[['Height']]), color='black', linewidth=3) # generated linear straight line
plt.scatter(X, y, color='red', marker='*') # Given dataset
plt.title("Height vs Weight")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```
